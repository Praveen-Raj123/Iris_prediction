```python
import pandas as pd
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
%matplotlib inline
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
data = pd.DataFrame(iris.data, columns = iris.feature_names)
data.head()
```

```python
data.shape
```

```python
data['target'] = iris.target
data.head()
```

```python
iris.target_names
```

```python
data[data.target==1].head()
```

```python
data['flower_name'] = iris.target_names[data['target']] 
data.head()
```

```python
data_0 =data[data.target==0]
data_1 =data[data.target==1]
data_2 =data[data.target==2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

plt.scatter(data_0['sepal length (cm)'], data_0['sepal width (cm)'], color = 'red', marker = 'o')
plt.scatter(data_1['sepal length (cm)'], data_1['sepal width (cm)'], color = 'blue', marker = '+')
plt.show()
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')

plt.scatter(data_0['petal length (cm)'], data_0['sepal width (cm)'], color = 'red', marker = 'o')
plt.scatter(data_1['petal length (cm)'], data_1['sepal width (cm)'], color = 'blue', marker = '+')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(['target','flower_name'], axis = 'columns')
y = data.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 42)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
